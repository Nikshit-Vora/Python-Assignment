Q1. How can you create a Bokeh plot using Python code?

Ans. Bokeh is a Python library for creating interactive visualizations for the web. Here's an example of how to create a simple Bokeh plot using Python code:

```python
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
```

```python
import numpy as np
from bokeh.plotting import figure, output_file, show

x = np.linspace(0, 10, 100)
y = np.sin(x)
output_file("plot.html")

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

p.line(x, y, line_width=2)

show(p)
```

Q2. What are glyphs in Bokeh, and how can you add them to a Bokeh plot? Explain with an example.

Ans. In Bokeh, a glyph is a visual shape that represents the data points in a plot. Glyphs can take many forms, including lines, circles, rectangles, and more. Here's an example of how to add a glyph to a Bokeh plot using Python code:

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import ColumnDataSource

# Output to a static HTML file
output_file("glyph.html")

# Define the data
x = [1, 2, 3, 4, 5]
y = [1, 4, 3, 2, 5]
colors = ["red", "green", "blue", "orange", "purple"]

# Create a ColumnDataSource object
source = ColumnDataSource(data=dict(x=x, y=y, color=colors))

# Create a new plot with a title and axis labels
p = figure(title="Glyph Example", x_axis_label='X-axis', y_axis_label='Y-axis')

# Add a circle glyph to the plot
p.circle('x', 'y', size=20, color='color', source=source)

# Show the plot
show(p)
```

Q3. How can you customize the appearance of a Bokeh plot, including the axes, title, and legend?

Ans. Bokeh provides a wide range of options for customizing the appearance of a plot, including the axes, title, and legend. Here are some examples of how to customize these elements using Python code:

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)
output_file("plot.html")

p = figure(title="Sine Wave", x_axis_label='x', y_axis_label='sin(x)')

p.line(x, y, line_width=2)

# customizing x axis..

p.xaxis.axis_label_text_color = "blue"
p.xaxis.axis_label_text_font_style = "italic"
p.xaxis.major_label_text_color = "red"
p.xaxis.major_label_orientation = "vertical"

# Customize the y-axis
p.yaxis.axis_label_text_color = "green"
p.yaxis.axis_label_text_font_style = "italic"
p.yaxis.major_label_text_color = "purple"
p.yaxis.major_label_orientation = "horizontal"

# Customize the title
p.title.text_font_size = "20pt"
p.title.text_color = "red"
p.title.align = "center"

show(p)

```

```python
x = [3,4,5,6,7]
y = [4,5,6,7,4]
output_file("line.html")
p = figure(title = "line graph")
p.scatter(x,y,size=20,fill_color="red",legend_label="Data Point")

# Customize the legend
p.legend.location = "top_left"
p.legend.label_text_font_size = "19pt"

show(p)
```

Q4. What is a Bokeh server, and how can you use it to create interactive plots that can be updated in real time?

Ans. Bokeh server is a feature of Bokeh that allows you to create interactive plots that can be updated in real time based on user input or streaming data. When using the Bokeh server, the plot is generated and served dynamically as a web application, and any changes made to the plot are immediately reflected in the user's browser without the need to reload the page.

To use the Bokeh server, you need to define your plot and data sources in Python, and then create a function that will update the plot based on user input or streaming data. This function is called a "callback" function and is typically defined using the @bokeh_app.route decorator.
