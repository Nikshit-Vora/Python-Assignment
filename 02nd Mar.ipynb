Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data plot a scatter plot.
Note: Also add title, xlabel, and ylabel to the plot.

Ans. A scatter plot is a type of data visualization that displays the relationship between two continuous variables by plotting them as a collection of points on a two-dimensional coordinate system. It is commonly used to examine the correlation or association between the variables and to detect any patterns, clusters, or outliers in the data.

```python
import numpy as np
np.random.seed(3)
x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))
```

```python
x
```

```python
y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(5,3))
plt.scatter(x,y,c="r",alpha=0.5)
plt.grid()
plt.xlabel("X data points")
plt.ylabel("Y data points")
plt.title("chart title : x v/s y data points")
plt.show()
```

Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

Ans. The subplot() function is used to create multiple plots in a single figure. It allows us to arrange multiple plots in a grid pattern, where each plot can have its own axes, labels, titles, and styles.

```python
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])

x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])

x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])

x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
```

```python
# plot1
plt.subplot(2,2,1)
plt.plot(x1,y1,c="orange")
plt.xlabel("X1")
plt.ylabel("Y1")
plt.title("Line 1")
plt.grid()

#plot2
plt.subplot(2,2,2)
plt.plot(x2,y2,c="red")
plt.xlabel("X2")
plt.ylabel("Y2")
plt.title("Line 2")
plt.grid()

#plot3
plt.subplot(2,2,3)
plt.plot(x3,y3,c="green")
plt.xlabel("X3")
plt.ylabel("Y3")
plt.title("Line 3")
plt.grid()

#plot4
plt.subplot(2,2,4)
plt.plot(x3,y3)
plt.xlabel("X4")
plt.ylabel("Y4")
plt.title("Line 4")
plt.grid()

plt.subplots_adjust(wspace=0.3, hspace=0.5)
plt.show()
```

Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.

Ans. A bar plot is a graph that represents categorical data with rectangular bars, where the height or length of the bars corresponds to the frequency or count of the data. 

It is used to compare and visualize the values of different categories or groups, show trends or changes over time, and highlight differences or similarities between the data.

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
plt.bar(company,profit)
plt.xlabel("Company")
plt.ylabel("Profit")
plt.title("Profit charts of various companies")
plt.grid()
plt.show()
```

```python
plt.barh(company,profit)
plt.ylabel("Company")
plt.xlabel("Profit")
plt.title("Profit charts of various companies")
plt.grid()
plt.show()
```

Ans. A box plot is a type of data visualization that displays the distribution of numerical data through their quartiles, median, and outliers. 

It is used to summarize and compare the center, spread, and skewness of different groups of data, and to identify any unusual or extreme values. Box plots can also be used to show the variability and symmetry of the data, and to detect any potential relationships or patterns among the variables.

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
data = [box1, box2]

fig = plt.figure(figsize=(5,2))
ax = fig.add_axes([0,0,1,1])

ax.boxplot(data)

plt.show()
```
